### Cleaning Data

```python
products_df[products_df.product_category_name.isna()]
```

```python
# Untuk kolom kategorikal
products_df['product_category_name'] = products_df['product_category_name'].fillna('uncategorized')
```

```python
# Untuk kolom length, description length, dan photos
numeric_cols = ['product_name_lenght', 'product_description_lenght', 'product_photos_qty']
products_df[numeric_cols] = products_df[numeric_cols].fillna(products_df[numeric_cols].median())
```

```python
# Untuk kolom dimensi fisik
dimension_cols = ['product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']
products_df[dimension_cols] = products_df[dimension_cols].interpolate(method='linear')
```

```python
products_df.isna().sum()
```

```python
order_df[order_df.order_approved_at.isna()]
```

```python
# Gunakan imputasi berbasis waktu: tambahkan waktu standar setelah order_purchase_timestamp (misal +1 jam)
order_df['order_purchase_timestamp'] = pd.to_datetime(order_df['order_purchase_timestamp'])
order_df['order_approved_at'] = pd.to_datetime(order_df['order_approved_at'])
def impute_approved_at(row):
    if pd.isnull(row['order_approved_at']):
        return row['order_purchase_timestamp'] + timedelta(hours=1)
    return row['order_approved_at']
order_df['order_approved_at'] = order_df.apply(impute_approved_at, axis=1)
```

```python
# Konversi semua kolom timestamp ke datetime
timestamp_columns = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]

for col in timestamp_columns:
    order_df[col] = pd.to_datetime(order_df[col])
def impute_carrier_date(row):
    if pd.isnull(row['order_delivered_carrier_date']) and pd.notnull(row['order_delivered_customer_date']):
        return row['order_delivered_customer_date'] - timedelta(days=2)
    return row['order_delivered_carrier_date']
order_df['order_delivered_carrier_date'] = order_df.apply(impute_carrier_date, axis=1)
```

```python
def handle_customer_delivery_date(row):
    if pd.isnull(row['order_delivered_customer_date']):
        if row['order_status'] == 'canceled':
            return None
        elif row['order_status'] in ['delivered', 'shipped']:
            if pd.notnull(row['order_estimated_delivery_date']):
                return row['order_estimated_delivery_date']
            elif pd.notnull(row['order_delivered_carrier_date']):
                return row['order_delivered_carrier_date'] + timedelta(days=2)
    return row['order_delivered_customer_date']

order_df['order_delivered_customer_date'] = order_df.apply(handle_customer_delivery_date, axis=1)
```

## Exploratory Data Analysis (EDA)

### Explore ...

```python
# Membuat dataframe untuk analisis pengiriman
delivery_df = order_df.merge(customers_df[['customer_id', 'customer_state']], 
                            on='customer_id', how='left')

# Menambahkan kolom-kolom analisis waktu
delivery_df['delivery_delay'] = (delivery_df['order_delivered_customer_date'] - 
                                delivery_df['order_estimated_delivery_date']).dt.days

delivery_df['actual_delivery_time'] = (delivery_df['order_delivered_customer_date'] - 
                                     delivery_df['order_purchase_timestamp']).dt.days

delivery_df['estimated_delivery_time'] = (delivery_df['order_estimated_delivery_date'] - 
                                        delivery_df['order_purchase_timestamp']).dt.days


# Kategorisasi keterlambatan
delivery_df['delivery_status'] = np.where(delivery_df['delivery_delay'] <= 0, 
                                        'On Time', 'Delayed')

# Membuat pivot table untuk analisis per state
state_performance = pd.pivot_table(delivery_df,
                                 index='customer_state',
                                 values=['delivery_delay', 'actual_delivery_time'],
                                 aggfunc={
                                     'delivery_delay': ['mean', 'count', 'std'],
                                     'actual_delivery_time': ['mean', 'min', 'max']
                                 }).round(2)

# Menghitung persentase keterlambatan per state
state_delay_percentage = delivery_df.groupby('customer_state').agg({
    'order_id': 'count',
    'delivery_status': lambda x: (x == 'Delayed').mean() * 100
}).round(2)
```

```python
state_performance.head()
```

```python
# Membuat dataframe untuk analisis produk
product_delivery_df = delivery_df.merge(order_items_df, on='order_id', how='left')\
                                .merge(products_df, on='product_id', how='left')
```

```python
# Kategorisasi produk berdasarkan dimensi
product_delivery_df['volume'] = (product_delivery_df['product_length_cm'] * 
                                product_delivery_df['product_height_cm'] * 
                                product_delivery_df['product_width_cm'])
```

```python
# Kategorisasi berat produk
def categorize_weight(weight):
    if weight <= 500:
        return 'Light'
    elif weight <= 2000:
        return 'Medium'
    else:
        return 'Heavy'

product_delivery_df['weight_category'] = product_delivery_df['product_weight_g'].apply(categorize_weight)
```

```python
# Pivot table untuk analisis kategori produk
category_performance = pd.pivot_table(product_delivery_df,
                                    index='product_category_name',
                                    values=['delivery_delay', 'volume', 'product_weight_g'],
                                    aggfunc={
                                        'delivery_delay': ['mean', 'count', 'std'],
                                        'volume': 'mean',
                                        'product_weight_g': 'mean'
                                    }).round(2)
```

```python
# Analisis status pengiriman berdasarkan karakteristik produk
weight_status_pivot = pd.pivot_table(product_delivery_df,
                                   index='weight_category',
                                   columns='order_status',
                                   values='order_id',
                                   aggfunc='count',
                                   fill_value=0)
```

```python
# Menghitung persentase keterlambatan berdasarkan kategori produk
category_delay_percentage = product_delivery_df.groupby('product_category_name').agg({
    'order_id': 'count',
    'delivery_status': lambda x: (x == 'Delayed').mean() * 100
}).round(2)
```

```python
category_performance.head()
```

```python
# Membuat dataframe untuk analisis waktu
delivery_df['order_month'] = delivery_df['order_purchase_timestamp'].dt.to_period('M')
delivery_df['order_weekday'] = delivery_df['order_purchase_timestamp'].dt.day_name()
```

```python
# Pivot table untuk analisis waktu
time_performance = pd.pivot_table(delivery_df,
                                index='order_month',
                                values='delivery_delay',
                                aggfunc=['mean', 'count', 'std']).round(2)
```

```python
# Analisis performa per hari dalam seminggu
weekday_performance = pd.pivot_table(delivery_df,
                                   index='order_weekday',
                                   values='delivery_delay',
                                   aggfunc=['mean', 'count', 'std']).round(2)
```

```python
weekday_performance.head()
```

```python
time_performance.head()
```

```python
state_performance_flat = state_performance.reset_index()
state_delay_percentage_flat = state_delay_percentage.reset_index()
category_performance_flat = category_performance.reset_index()
category_delay_percentage_flat = category_delay_percentage.reset_index()
time_performance_flat = time_performance.reset_index()
weekday_performance_flat = weekday_performance.reset_index()

# Tambahkan kolom sumber analisis untuk identifikasi
delivery_df['analysis_type'] = 'delivery'
state_performance_flat['analysis_type'] = 'state_performance'
state_delay_percentage_flat['analysis_type'] = 'state_delay_percentage'
category_performance_flat['analysis_type'] = 'category_performance'
category_delay_percentage_flat['analysis_type'] = 'category_delay_percentage'
time_performance_flat['analysis_type'] = 'time_performance'
weekday_performance_flat['analysis_type'] = 'weekday_performance'

# Gabungkan semua DataFrame menjadi satu
all_df = pd.concat([
    delivery_df,
    state_performance_flat,
    state_delay_percentage_flat,
    category_performance_flat,
    category_delay_percentage_flat,
    time_performance_flat,
    weekday_performance_flat
], ignore_index=True, sort=False)

# Simpan sebagai file CSV
all_df.to_csv('all_analysis_data.csv', index=False)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
plt.figure(figsize=(15, 10))

# Distribusi Keterlambatan per State
plt.subplot(2, 2, 1)
state_delays = delivery_df.groupby('customer_state')['delivery_delay'].mean().sort_values(ascending=False)
sns.barplot(x=state_delays.index, y=state_delays.values)
plt.title('Rata-rata Keterlambatan per State')
plt.xticks(rotation=45)
plt.ylabel('Rata-rata Keterlambatan (Hari)')

# Persentase Keterlambatan
plt.subplot(2, 2, 2)
on_time = (delivery_df['delivery_delay'] <= 0).sum()
delayed = (delivery_df['delivery_delay'] > 0).sum()
plt.pie([on_time, delayed], 
        labels=['Tepat Waktu', 'Terlambat'], 
        autopct='%1.1f%%',
        colors=['lightgreen', 'lightcoral'])
plt.title('Proporsi Ketepatan Waktu Pengiriman')
```

### Pertanyaan 2:

```python
plt.figure(figsize=(15, 10))

# Top 10 Kategori dengan Keterlambatan Tertinggi
plt.subplot(2, 2, 2)
category_delays = product_delivery_df.groupby('product_category_name')['delivery_delay'].mean()
top_10_delays = category_delays.nlargest(10)
sns.barplot(x=top_10_delays.values, y=top_10_delays.index)
plt.title('10 Kategori Produk dengan Keterlambatan Tertinggi')
plt.xlabel('Rata-rata Keterlambatan (Hari)')

# Heatmap Korelasi
plt.subplot(2, 2, 4)
correlation_features = ['delivery_delay', 'product_weight_g', 'product_length_cm', 
                       'product_height_cm', 'product_width_cm', 'volume']
correlation = product_delivery_df[correlation_features].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0)
plt.title('Korelasi antara Karakteristik Produk dan Keterlambatan')

plt.tight_layout()
plt.show()
```
